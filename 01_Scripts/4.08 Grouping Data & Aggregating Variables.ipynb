### (5.) The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Descriptive Analyses
ords_prods_merge.groupby('user_id').agg({'prices' : ['mean','min','max']}).head(15)
```

```python
# Group by user_id, aggregate prices and trandform the mean price to the spending column
ords_prods_merge['spending'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
# Create spending flags by loc functions
ords_prods_merge.loc[ords_prods_merge['spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check the result of loc functions
ords_prods_merge[['user_id','order_number','prices','spending','spending_flag']].head(20)
```

### (6.) In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Group by user_id and aggreagte days_since_prior_order in median value
ords_prods_merge.groupby('user_id').agg({'days_since_prior_order': ['median']}).head(5)
```

```python
# Group by user_id, aggreagte days_since_prior_order in median value, and transform the value to frequency column
ords_prods_merge['frequency'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
# Create frequency flags by loc functions
ords_prods_merge.loc[ords_prods_merge['frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['frequency'] > 10) & (ords_prods_merge['frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check the result for loc functions
ords_prods_merge[['user_id','frequency','frequency_flag']].head(10)
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[ords_prods_merge['frequency_flag'].isnull() == True]
```

Missing values come from user_id 159838, who don't have previous order. We can filter out these rows since we just want to categorize customers in three categories based on the frequency. 

```python
ords_prods_merge_noNaN=  ords_prods_merge[ords_prods_merge['frequency_flag'].isnull() == False]
```
